## Penguin Species Classification – Machine Learning Project
This project aims to build a robust classification system to accurately identify the species of penguins based on physical characteristics and demographic attributes, using the Palmer Penguins dataset. It provides a complete machine learning pipeline—from data preparation to model evaluation—with integrated visualizations and clear insights.

This dataset contain data for 344 penguins. There are 3 different species of penguins in this dataset, collected from 3 islands in the Palmer Archipelago, Antarctica. The dataset consists of 7 columns, 3 of them contain categorical features and 4 of them numerical features:

species: penguin species (Chinstrap, Adélie, or Gentoo)

culmen_length_mm: culmen length (mm)

culmen_depth_mm: culmen depth (mm)

flipper_length_mm: flipper length (mm)

body_mass_g: body mass (g)

island: island name (Dream, Torgersen, or Biscoe) in the Palmer Archipelago (Antarctica)

sex: penguin sex

The task is to predict the species (target) using the featues above and a number of different machine learning algorithm (KNN classifier, support vector machine, decision tree classifier, and random forrest classifier). 

What is culmen?
The upper margin of the beak or bill is referred to as the culmen and the measurement is taken using calipers with one jaw at the tip of the upper mandible and the other at base of the skull or the first feathers depending on the standard chosen.

### Data Cleaning & Preprocessing
Missing values in numerical features (e.g., culmen length, body mass) were filled using mean imputation.

Missing and invalid categorical entries (e.g., . in the sex column) were corrected and filled using the mode.

Categorical variables (species, island, sex) were encoded using LabelEncoder.

```python
# Import Libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.metrics import accuracy_score
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
```

```python
# Load Dataset
df = pd.read_csv("data/penguins_size.csv")
```

```python
df.head(5)
```

```python
df.info()
```

```python
df.isnull().sum()
```

There are 5 columns having missing values: `culmen_length_mm`, `culmen_depth_mm`, `flipper_length_mm`, `body_mass_g` and `sex`. As these features `culmen_length_mm`, `culmen_depth_mm`, `flipper_length_mm`, `body_mass_g` are numerical, the missing value will be replaced by the mean. `sex` is categorical fetures so the missing values will be replaced by the mode.

```python
# Fill missing numerical values with mean
for col in ['culmen_length_mm', 'culmen_depth_mm', 'flipper_length_mm', 'body_mass_g']:
    df[col].fillna(df[col].mean(), inplace=True)

# Fill missing categorical values with mode
df['sex'].fillna(df['sex'].mode()[0], inplace=True)
```

```python
# Value Counts
print("Species Distribution:\\n", df['species'].value_counts())
print("\\nIsland Distribution:\\n", df['island'].value_counts())
print("\\nSex Distribution:\\n", df['sex'].value_counts())
```

```python
# Clean Data
df['sex'] = df['sex'].replace('.', 'FEMALE')
```

```python
# Check data
df.isnull().sum()
```

The dataset comprises three penguin species: Adelie (152 samples), Gentoo (124 samples), and Chinstrap (68 samples), with Adelie being the most common and Chinstrap the least represented.

Examining the numerical feature distributions (Figure 2), Gentoo penguins stand out as significantly heavier, with longer flippers and lower culmen depth compared to the other species. In contrast, Adelie penguins tend to have shorter culmen lengths, lighter body mass, and shorter flipper length, while Chinstrap penguins fall somewhere in between, overlapping partially with both Adelie and Gentoo in terms of physical measurements.

The correlation matrix (Figure 3) reveals a strong positive correlation between body mass and flipper length (approximately 0.87), while other features show moderate to weak correlations. These insights suggest that a few key physical attributes—particularly body mass and flipper length—are likely to be highly predictive in distinguishing between penguin species.

Analyzing the island distribution (Figure 4) reveals that Chinstrap penguins are found exclusively on Dream Island, Gentoo are only present on Biscoe Island, while Adelie penguins inhabit all three islands—Torgersen, Biscoe, and Dream. The sex distribution across the dataset is relatively balanced, although slight variations exist between species.

```python
# Encode Categorical Data
df_encoded = df.copy()
label_encoders = {}
for col in ['species', 'island', 'sex']:
    le = LabelEncoder()
    df_encoded[col] = le.fit_transform(df_encoded[col])
    label_encoders[col] = le
```

### Model Training
Four models were trained using an 80/20 train-test split:

KNeighborsClassifier (n_neighbors=6)

SVC(kernel='poly', random_state=100)

DecisionTreeClassifier(random_state=100)

RandomForestClassifier(n_estimators=100, random_state=100)

Model training was conducted before and after applying StandardScaler, to evaluate the effect of feature scaling on model performance.

```python
# Train/Test Split (80/20)
X = df_encoded.drop('species', axis=1)
y = df_encoded['species']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=100)
```

```python
# Train Models (Unscaled)
models_unscaled = {
    "KNN (k=6)": KNeighborsClassifier(n_neighbors=6),
    "SVM (poly)": SVC(kernel='poly', random_state=100),
    "Decision Tree": DecisionTreeClassifier(random_state=100),
    "Random Forest": RandomForestClassifier(n_estimators=100, random_state=100)
}

results_unscaled = {}
for name, model in models_unscaled.items():
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    results_unscaled[name] = accuracy_score(y_test, y_pred)
```

```python
# Scale Features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
X_train_scaled, X_test_scaled, y_train_scaled, y_test_scaled = train_test_split(
    X_scaled, y, test_size=0.2, random_state=100
)
```

```python
# Train Models (Scaled)
models_scaled = {
    "KNN (k=6)": KNeighborsClassifier(n_neighbors=6),
    "SVM (poly)": SVC(kernel='poly', random_state=100),
    "Decision Tree": DecisionTreeClassifier(random_state=100),
    "Random Forest": RandomForestClassifier(n_estimators=100, random_state=100)
}

results_scaled = {}
for name, model in models_scaled.items():
    model.fit(X_train_scaled, y_train_scaled)
    y_pred = model.predict(X_test_scaled)
    results_scaled[name] = accuracy_score(y_test_scaled, y_pred)
```

```python
# Visualization of Accuracy Comparison with values on bars
labels = list(results_unscaled.keys())
x = np.arange(len(labels))
width = 0.35

fig, ax = plt.subplots(figsize=(10, 6))
bars1 = ax.bar(x - width/2, [results_unscaled[l] for l in labels], width, label='Unscaled')
bars2 = ax.bar(x + width/2, [results_scaled[l] for l in labels], width, label='Scaled')

# Add accuracy values on top of each bar
for bar in bars1 + bars2:
    height = bar.get_height()
    ax.annotate(f'{height:.2f}',
                xy=(bar.get_x() + bar.get_width() / 2, height),
                xytext=(0, 3),  # 3 points vertical offset
                textcoords="offset points",
                ha='center', va='bottom', fontsize=10)

# Customize chart
ax.set_ylabel('Accuracy')
ax.set_title('Figure 5: Accuracy Comparison (Unscaled vs Scaled Features)')
ax.set_xticks(x)
ax.set_xticklabels(labels, rotation=45)
ax.legend()

plt.tight_layout()

# Save the figure
plt.savefig("accuracy_comparison_scaled_vs_unscaled.png", dpi=150, bbox_inches='tight')

# Show the plot
plt.show()
```

Scaling the dataset significantly improved k-Nearest Neighbors (kNN) and Support Vector Classifier (SVC) accuracy from 0.78 and 0.74, respectively to a perfect 1.0. This can be explained that scaling may have a significant impact on kNN. The algorithm relies on distances between data points, so scaling can lead to more accurate results. Similarly, SVC can also benefit from scaling. Scaling ensures that all features have a similar influence on the decision boundary.

In contrast, Decision Tree and Random Forest models, which are less sensitive to feature scaling, maintained their high accuracy scores 0.99 and 1 accordingly even without scaling. Decision Trees are not affected by feature scaling since they make decisions based on feature thresholds. Random Forests, like Decision Trees, are not sensitive to feature scaling. Therefore, scaling may not have a substantial impact on Decision Tree and Random Forest accuracy.
